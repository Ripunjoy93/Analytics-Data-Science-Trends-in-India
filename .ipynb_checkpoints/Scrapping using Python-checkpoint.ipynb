```python
import pandas as pd
```

```python
u_cols = ['user_id', 'age', 'sex', 'occupation', 'zip_code']

users = pd.read_csv(
    'http://files.grouplens.org/datasets/movielens/ml-100k/u.user', 
    sep='|', names=u_cols)

users.head()
```

```python
users.size
```

```python
# age>40 and male
male = users[users.sex == "M"]
male40 = male[male.age > 40]
male40.head()
```

```python
# alternative
male40 = users[(users.age > 40) & (users.sex=="M") ]
male40.head()
```

```python
# mean age of female progrm
femaleP = users[(users.sex == "F") & (users.occupation == "programmer")]

print (femaleP.age.mean())
```

```python
# why ?
print (femaleP.age.mean)
```

```python
r_cols = ['user_id', 'movie_id', 'rating', 'unix_timestamp']
ratings = pd.read_csv(
    'http://files.grouplens.org/datasets/movielens/ml-100k/u.data', 
    sep='\t', names=r_cols)

ratings.head() 
```

```python
movieR = ratings['rating'].groupby(ratings['movie_id'])
avgmR = movieR.mean()
print (avgmR)
```
